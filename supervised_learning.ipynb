```python
car_df=data.drop('car name',axis=1)
car_df
```

```python
car_df['origin']=car_df['origin'].replace({1:'america',2:'europe',3:'asia'})
car_df
```

```python
car_df=pd.get_dummies(car_df,columns=['origin'])
car_df
```

DEALING WITH MISSING VALUES

```python
car_df.describe()
```

```python
car_df.info()
```

```python
car_df.dtypes
```

```python
car_df['hp']=car_df['hp'].replace('?',np.nan)
car_df['hp']=car_df['hp'].astype(float)
```

```python
car_df.dtypes
```

```python
car_df.isnull().sum()
```

```python
median1=car_df['hp'].mean()
```

```python
car_df['hp']=car_df['hp'].replace(np.nan,median1)
```

split data

```python
x=car_df.drop(['mpg'],axis=1)
y=car_df[['mpg']]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=1)
```

FIT LINEAR MODEL

```python
reg_model=LinearRegression()
reg_model.fit(x_train,y_train)
```

```python
reg_model.score(x_train,y_train)
```

```python
reg_model.score(x_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

poly=PolynomialFeatures(degree=2,interaction_only=True)
x_train2=poly.fit_transform(x_train)
x_test2=poly.fit_transform(x_test)

poly_clf=linear_model.LinearRegression()

poly_clf.fit(x_train2,y_train)

print(poly_clf.score(x_train2,y_train))
```

```python
print(poly_clf.score(x_test2,y_test))
```
