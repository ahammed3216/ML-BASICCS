```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR=Q3-Q1
    lower_range=Q1-(1.5*IQR)
    upper_range=Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
low_time,upper_time=remove_outlier(data["lead_time"])
data["lead_time"]=np.where(data["lead_time"]>upper_time,upper_time,data["lead_time"])
data["lead_time"]=np.where(data["lead_time"]<low_time,low_time,data["lead_time"])
```
